```python
# Importeer de Pandas bibliotheek voor data-analyse
# Importeer Matplotlib voor het maken van grafieken
import pandas as pd
import matplotlib.pyplot as plt
```

```python
knmi = pd.read_csv("knmi.csv")
knmi
```

```python
### Opdracht 3: Herbenoemen van de kolomnamen naar het Nederlands met .rename() ###
```

```python
knmi = knmi.rename(columns={"MAR":"MRT", "MAY":"MEI","OCT":"OKT","YEAR":"JAAR"})
knmi
```

```python
### Opdracht 4: Verander de graden Celcius van huidige tiende graad naar graden ###
```

```python
knmi.iloc[:, 2:] = knmi.iloc[:, 2:]/10
knmi.head()
```

```python
### Opdracht 5: Print de hoogste gemiddelde jaartemperatuur en het desbetreffende jaar uit.  ###
```

```python
highest_mean_temp = knmi.loc[knmi["JAAR"].idxmax(),["YYYY","JAAR"]]
highest_mean_temp
print(f"Hoogste jaargemiddelde: {highest_mean_temp["JAAR"]}°C")
print(f"Dit was in jaar: {int(highest_mean_temp["YYYY"])}")
```

```python
### Opdracht 6:  Maak een grafiek van de gemiddelde jaartemperatuur ###
```

```python
# Importeer Matplotlib voor het maken van grafieken
import matplotlib.pyplot as plt

# Voor het maken van een grafiek gebruiken we alleen de data die we willen gebruiken.
# Maak een subset van het knmi DataFrame...
gem_jaar_temp = knmi[["YYYY","JAAR"]].dropna().reset_index()

# Toon de eerste vijf rijen van deze subset
gem_jaar_temp.head()
```

```python
# Maak de grafiek.
plt.figure(figsize=(10,8))
plt.plot(gem_jaar_temp["YYYY"], gem_jaar_temp["JAAR"], marker="o", label="Temperatuur")

plt.grid(axis="y")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Meer uitgebreide grafiek
plt.figure(figsize=(10,8))
plt.plot(gem_jaar_temp["YYYY"], gem_jaar_temp["JAAR"], marker="o", label="Temperatuur")

# Trendlijn
z = np.polyfit(gem_jaar_temp["YYYY"], gem_jaar_temp['JAAR'], 1)
p = np.poly1d(z)
plt.plot(gem_jaar_temp["YYYY"], p(gem_jaar_temp["YYYY"]), linestyle='--', color='red', label='Trendline')

plt.title("Gemiddelde Jaartemperatuur 1900-2023", color="black", fontsize=16, pad=20)
plt.xlabel("Jaar")
plt.ylabel("Temperatuur")
plt.grid(axis="y")
plt.legend()

plt.tight_layout()
plt.show()
```
